```python
import os
import pandas as pd
```

```python
dfs = []

for f in os.listdir("with_legacy_format"):
    if f.endswith(".csv"):
        df = pd.read_csv(f"with_legacy_format/{f}", sep=",", dtype={"energy_consumed": float, "emissions": float})
        dfs.append(df)

with_legacy_format_dfs = pd.concat(dfs)
with_legacy_format_dfs.head()

with_legacy_format_average_df = with_legacy_format_dfs.groupby("project_name")[["energy_consumed", "emissions"]].mean().reset_index()
with_legacy_format_average_df.head(20)
```

```python
print(with_legacy_format_average_df.to_latex())
```

```python
dfs = []

for f in os.listdir("without_legacy_format"):
    if f.endswith(".csv"):
        df = pd.read_csv(f"without_legacy_format/{f}", sep=",", dtype={"energy_consumed": float, "emissions": float})
        dfs.append(df)

without_legacy_format_dfs = pd.concat(dfs)
without_legacy_format_dfs.head()

without_legacy_format_average_df = without_legacy_format_dfs.groupby("project_name")[["energy_consumed", "emissions"]].mean().reset_index()
without_legacy_format_average_df.head(20)
```

```python
print(without_legacy_format_average_df.to_latex())
```

```python
difference = with_legacy_format_average_df[['project_name']].copy()
difference[["energy_consumed", "emissions"]] = with_legacy_format_average_df[["energy_consumed", "emissions"]] - without_legacy_format_average_df[["energy_consumed", "emissions"]]
difference.head(20)
```

```python
print(difference.to_latex())
```

```python
difference_in_percentage = with_legacy_format_average_df[['project_name']].copy()
difference_in_percentage[["energy_consumed", "emissions"]] = ((with_legacy_format_average_df[["energy_consumed", "emissions"]] - without_legacy_format_average_df[["energy_consumed", "emissions"]]) / with_legacy_format_average_df[["energy_consumed", "emissions"]]) * 100
difference_in_percentage = difference_in_percentage.rename(
    columns={
        "energy_consumed": "energy_consumed_percentage_change",
        "emissions": "emissions_percentage_change"
    }
)
difference_in_percentage["energy_consumed_percentage_change"] = difference_in_percentage["energy_consumed_percentage_change"].round(1).astype(str) + "%"
difference_in_percentage["emissions_percentage_change"] = difference_in_percentage["emissions_percentage_change"].round(1).astype(str) + "%"
difference_in_percentage.head(20)
```

```python
print(difference_in_percentage.to_latex())
```
